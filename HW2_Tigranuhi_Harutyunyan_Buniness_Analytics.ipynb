```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cross_validation import train_test_split
```

```python
# reading the data
data = pd.read_excel("IBM_HW2.xlsx",)
data.head()
```

```python
#dropping the columns that we consider not revelant for our analysis
#we drop EmployeeCount,Over18,StandardHours columns,as the values are the same and there is no variance,so there will be no effect on desicion tree
data["EmployeeCount"].var()
```

```python
data["Over18"].var()
```

```python
data["StandardHours"].var()
```

```python
#dropping EmployeeCount
data = data.drop(["EmployeeCount"],axis=1)
data.head()
```

```python
#dropping Over18
data = data.drop(["Over18"],axis=1)
data.head()
```

```python
#dropping StandardHours
data = data.drop(["StandardHours"],axis=1)
data.head()
```

```python
#as we have Hourly,Daily and Monthly Rates, we need to include the one which has the highest varinace
data["HourlyRate"].var()
```

```python
data["DailyRate"].var()
```

```python
data["MonthlyRate"].var()
```

```python
#MonthlyRate has the highest variance, so we drop HourlyRate and DailyRate
data = data.drop(["HourlyRate"],axis=1)
data.head()
```

```python
data = data.drop(["DailyRate"],axis=1)
data.head()
```

```python
#dropping EmoloyeeNumber as it is not relevent information for analysis, since each employee has its number(ID)
data = data.drop(["EmployeeNumber"],axis=1)
data.head()
```

```python
#categorising the Gender and Marital Status variables to get integers to work easier
data["Gender"] = data["Gender"].astype('category').cat.reorder_categories(['Male', 'Female']).cat.codes
data.head()
```

```python
data["MaritalStatus"] = data["MaritalStatus"].astype('category').cat.reorder_categories(['Single', 'Married','Divorced']).cat.codes
data.head()
```

```python
#getting dummy variables for each not integers in our dataset(Department,EducationField,JobRole)
department = pd.get_dummies(data["Department"])
department.head()
```

```python
#dropping column Human Recources to avoid overlaping problem
new_department = department.drop(["Human Resources"],axis=1)
new_department.head()
```

```python
#droping the "old" column
data = data.drop(["Department"],axis=1)
data.head()
```

```python
edu = pd.get_dummies(data["EducationField"])
edu.head()
```

```python
new_edu = edu.drop(["Human Resources"],axis=1)
new_edu.head()
```

```python
data = data.drop(["EducationField"],axis=1)
data.head()
```

```python
job_role = pd.get_dummies(data["JobRole"])
job_role.head()
```

```python
new_jobrole = job_role.drop(["Human Resources"],axis=1)
new_jobrole.head()
```

```python
data = data.drop(["JobRole"],axis=1)
data.head()
```

```python
data = data.join(new_jobrole)
data.head()
```

```python
# learning the percentage of leavers
data["Attrition"].value_counts()/len(data)*100
```

```python
# importing the model
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=50)
```

```python
# setting the inputs and output
# inputs are everything except the column left
# output is the column left
inputs = data.iloc[:,1:]
output = data["Attrition"]
```

```python
# fitting the model
model.fit(inputs,output)
```

```python
model.score(inputs,output)*100
```

```python
#importing the library for exporting tree visualization code
# exporting the tree graph and then visualizing in webgraphviz.com
from sklearn.tree import export_graphviz
export_graphviz(model,"tree!.dot")
```
